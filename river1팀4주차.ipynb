```python
#미션 1
def make_comma(num):
  num_str = str(num) #입력받은 숫자를 문자열로 변환
  num_length = len(num_str) #문자열의 길이를 num_length 변수에 저장

  if num_length <= 3: #문자열의 길이가 3이하이면 쉼표를 붙일 필요가 없으므로 문자열 그대로 반환
    return num_str

  reversed_num_str = num_str[::-1] # 숫자를 거꾸로 탐색하기 위해 reversed_num_str에 거꾸로된 문자열을 저장
  formatted_num = ""

  for i in range(num_length):
    if i > 0 and i % 3 == 0: #숫자가 1234 라면 reversed_num_str에는 4321이 저장되고 쉼표가 들어갈 인덱스는 3의 배수가 됨
      formatted_num += ","  # 그러나 0도 3으로 나누면 0 이므로 0보다 큰 인덱스에서 3의 배수에 쉼표가 들어가면됨
                            # if문 조건을 만족시키지 않을 때 -> formatted_num 에다 reversed_num_str 복사
                            # if문 조건을 만족시킬 때(=쉼표가 들어가야될 때) -> "," 을 추가해줌
    formatted_num += reversed_num_str[i]

  return formatted_num[::-1] # formatted_num 에는 432,1 이 저장되있으므로 formatted_num 을 다시 거꾸로 뒤집어 반환한다

while True: #숫자를 입력받을때 까지 입력을 요구하는 반복문
  user_input = input("숫자를 입력하세요: ")

  if user_input.isdigit():
    user_input = int(user_input)
    make_comma_output = make_comma(user_input)
    print(f"포맷팅된 결과: {make_comma_output}")
    break
  else:
    print("숫자를 입력해주세요.")

```

```python
#미션2
def count_word(text, word):
  count = text.count(word) #text에 word가 들어가 있는지 확인하기
  print(count)

def save_to_txt(text, filename):
  with open(filename, 'w', encoding='utf-8') as file:
    file.write(text)


a ="""안녕하세요.
반갑습니다. 파이썬 공부는 정말 재밌습니다.
"""
count_word(a, "습니다")


filename = "output.txt"
save_to_txt(a, filename)
```

```python
#미션3
# 김갑,123456789
# 이을,010-1234-5678
# 박병,010-5678-111
# 최정,111-1111-1111
# 정무,010-3333-3333

def check_guest_book(guest_book):
  lines = guest_book.split("\n")
  wrong_guest = []

  for line in lines:
    name, phone_number = line.split(",")

    if phone_number.startswith("010") and len(phone_number) == 13 and '-' in phone_number:
      # 전화번호를 '-'로 나누어 세 그룹으로 나눔.
      groups = phone_number.split("-")

      if len(groups[1]) == 4 and len(groups[2]) == 4:
        if groups[1].isdigit() and groups[2].isdigit():
          continue  # 올바른 형식을 만족하는 경우 다음 참석자 검사로 넘어감.

    wrong_guest.append((name, phone_number)) # append를 이용해 잘못입력된 참석자 정보를 저장함

  return wrong_guest

def wrong_guest_book(guest_book):
  wrong_guest = check_guest_book(guest_book)

  for name, phone_number in wrong_guest:
    print(f"잘못 쓴 사람: {name}")
    print(f"잘못 쓴 번호: {phone_number}\n")

guest_book = """김갑,123456789
이을,010-1234-5678
박병,010-5678-111
최정,111-1111-1111
정무,010-3333-3333"""
wrong_guest_book(guest_book)

```
