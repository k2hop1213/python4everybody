```python
def king(korea_king, chosun_king):
  #쉼표를 기준으로 분할하기
  korea_king = korea_king.split(',')
  chosun_king = chosun_king.split(',')
  #공통된 부분 저장하기
  set_king = list(set(korea_king) & set(chosun_king))

  return set_king


korea_king = "태조,혜종,정종,광종,경종,성종,목종,현종,덕종,정종,문종,순종,선종,헌종,숙종,예종,인종,의종,명종,신종,희종,강종,고종,원조,충렬왕,충선왕,충숙왕,충혜왕,충목왕,충정왕,공민왕,우왕,창왕,공양왕"
chosun_king = "태조,정종,태종,세종,문종,단종,세조,예종,성종,연산군,중종,인종,명종,선조,광해군,인조,효종,현종,숙종,경종,영조,정조,순조,헌종,철종,고종,순종"
king_names = king(korea_king, chosun_king)
for name in king_names:
  print("조선과 고려에 모두 있는 왕 이름:", name)

print(f"조선과 고려에 모두 있는 왕 이름은 총 {len(king_names)}개 입니다")

```

```python
def sales_management(member_names, member_records):
  average_scores = []
  #각각의 평균을 저장해서 리스트에 저장
  for records in member_records:
    average_score = sum(records) / len(records)
    average_scores.append(average_score)
  #내림차순으로 정렬
  sorted_indices = sorted(range(len(average_scores)), key=lambda k: average_scores[k], reverse=True)

  #순위를 나눠서 보너스 대상자와 면담대상자 선정
  bonus_indices = sorted_indices[:2]
  counseling_indices = sorted_indices[-2:]

  bonus = [member_names[i] for i in bonus_indices if average_scores[i] > 5]
  counseling = [member_names[i] for i in counseling_indices if average_scores[i] <= 3]


  for name in bonus:
    print(f"보너스 대상자 {name}")

  for name in counseling:
    print(f"면담 대상자 {name}")


member_names = ["갑돌이", "갑순이", "을돌이", "을순이", "병돌이", "병순이"]
member_records = [
      [4, 5, 3, 5, 6, 5, 3, 4, 1, 3, 4, 5],
      [2, 3, 4, 3, 1, 2, 0, 3, 2, 5, 7, 2],
      [1, 3, 0, 3, 3, 4, 5, 6, 7, 2, 2, 1],
      [3, 2, 9, 2, 3, 5, 6, 6, 4, 6, 9, 9],
      [8, 7, 7, 5, 6, 7, 5, 8, 8, 6, 10, 9],
      [7, 8, 4, 9, 5, 10, 3, 3, 2, 2, 1, 3]
]

sales_management(member_names, member_records)


```

```python
def stock_profit(stocks_info, sells):
    stocks_list = stocks_info.split(',')
    stock_data = []

    for stock in stocks_list:
        name, quantity, purchase_price = stock.split('/')
        quantity = int(quantity)
        purchase_price = int(purchase_price)
        sell_price = sells.pop(0)
        profit_percent = ((sell_price - purchase_price) / purchase_price) * 100
        stock_data.append((name, profit_percent))

    sorted_stocks = sorted(stock_data, key=lambda x: x[1], reverse=True)

    for stock in sorted_stocks:
        name, profit_percent = stock
        print(f"{name}의 수익률 {profit_percent:.2f}")


stocks = "삼성전자/10/85000,카카오/15/130000,LG화학/3/820000,NAVER/5/420000"
sells = [82000, 160000, 835000, 410000]
stock_profit(stocks, sells)



```

```python
def good_customer(info):
    info_list = info.split(',')
    member_info = {
        '아이디': [],
        '나이': [],
        '전화번호': [],
        '성별': [],
        '지역': [],
        '구매횟수': []
    }

    vip_members = []

    for i in range(0, len(info_list), 6):
        member_info['아이디'].append(info_list[i])
        member_info['나이'].append(info_list[i+1])
        member_info['전화번호'].append(info_list[i+2] if info_list[i+2] != 'x' else '000-0000-0000')
        member_info['성별'].append(info_list[i+3])
        member_info['지역'].append(info_list[i+4])
        member_info['구매횟수'].append(int(info_list[i+5]))

    for i, purchases in enumerate(member_info['구매횟수']):
        if purchases >= 8 and member_info['전화번호'][i] != '000-0000-0000':
            vip_members.append(i)

    vip_info = {key: member_info[key] for key in member_info}

    for i in vip_members:
        print(f"할인 쿠폰을 받을 회원정보 아이디:{vip_info['아이디'][i]}, 나이:{vip_info['나이'][i]}, 전화번호:{vip_info['전화번호'][i]}, 성별:{vip_info['성별'][i]}, 지역:{vip_info['지역'][i]}, 구매횟수: {vip_info['구매횟수'][i]}")

info = "abc,21세,010-1234-5678,남자,서울,5,cdb,25세,x,남자,서울,4,bbc,30세,010-2222-3333,여자,서울,3,ccb,29세,x,여자,경기,9,dab,26세,x,남자,인천,8,aab,23세,010-3333-1111,여자,경기,10"
good_customer(info)


```
