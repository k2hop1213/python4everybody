```python
#미션 1
def grader(name,score) :
    print("학생이름:",name)
    print("점수:",score)

    if score >= 95 and score <= 100:
        print("학점:A+")
    elif score >= 90 :
        print("학점:A")
    elif score >= 85 :
        print("학점:B+")
    elif score >= 80 :
        print("학점:B")
    elif score >= 75 :
        print("학점:C+")
    elif score >= 70 :
        print("학점:C")
    elif score >= 65 :
        print("학점:D+")
    elif score >= 60 :
        print("학점:D")
    else :
        print("학점:F")

grader("이호창",99)
```

```python
#미션 2
def yearly_payment(monthly_payment):
    print("세전 연봉",monthly_payment*12,"만원")
    if monthly_payment*12 <= 1200 :
        print("세후 연봉:",monthly_payment*12 - ((monthly_payment*12)*(6/100)),"만원")
    elif monthly_payment*12 <= 4600 :
        print("세후 연봉:",monthly_payment*12 - ((monthly_payment*12)*(15/100)),"만원")
    elif monthly_payment*12 <= 8800 :
        print("세후 연봉:",monthly_payment*12 - ((monthly_payment*12)*(24/100)),"만원")
    elif monthly_payment*12 <= 15000 :
        print("세후 연봉:",monthly_payment*12 - ((monthly_payment*12)*(35/100)),"만원")
    elif monthly_payment*12 <= 30000 :
        print("세후 연봉:",monthly_payment*12 - ((monthly_payment*12)*(38/100)),"만원")
    elif monthly_payment*12 <= 50000 :
        print("세후 연봉:",monthly_payment*12 - ((monthly_payment*12)*(40/100)),"만원")
    else :
        print("세후 연봉:",monthly_payment*12 - ((monthly_payment*12)*(42/100)),"만원")

monthly_payment=300
yearly_payment(monthly_payment)
```

```python
#미션 3
# 지불 방식이 현금과 카드로 나뉘니 조건문으로 두가지 경우를 나누고
# 그 안에서 나이에 따라 조건문을 만들어 요금을 안내해준다.
def bus_fare(age,pay_type):
    print("나이:",age,"세")
    if pay_type == "현금" :
        print("지불유형:",pay_type)
        if age < 8 or age >= 75 :
            print("버스요금:무료")
        elif age >= 8 and age < 14 :
            print("버스요금:450원")
        elif age >= 14 and age < 20 :
            print("버스요금:1000원")
        else :
            print("버스요금:1300원")
    elif pay_type == "카드" :
        print("지불유형:",pay_type)
        if age < 8 or age >= 75 :
            print("버스요금:무료")
        elif age >= 8 and age < 14 :
            print("버스요금:450원")
        elif age >= 14 and age < 20 :
            print("버스요금:720원")
        else :
            print("버스요금:1200원")

bus_fare(30,"현금")
```

```python
#미션 4
import random

def rcp(my):
    computer_num = random.randint(0, 2) #computer_num에는 0에서 2의 정수가 랜덤하게 들어가게된다.
    computer = "가위" if computer_num == 0 else "바위" if computer_num == 1 else "보" #나온 정수에 따라  computer에 가위 바위 보 중 하나를 할당한다

    print("나:", my)
    print("컴퓨터:", computer)

    if my == computer:
        print("비겼습니다")
        my = input("가위, 바위, 보 중에 입력하세요: ") #비겼을 경우에는 승부를 가리기위해 가위 바위 보를 한 번 더 한다.
        rcp(my)
    elif (my == "가위" and computer == "보") or \
         (my == "바위" and computer == "가위") or \
         (my == "보" and computer == "바위"):
        print("내가 이겼다!")
    else:
        print("컴퓨터 승리!")

my = input("가위, 바위, 보 중에 입력하세요: ")
rcp(my)



```
