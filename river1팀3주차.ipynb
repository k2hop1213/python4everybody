```python
#미션 1

def gugudan(number):
  # 숫자에 해당하는 구구단 출력
  print(f"{number}단")
  i = 1
  while i <= 9:
    # 숫자와 i를 곱한 결과를 value 변수에 저장
    value = number * i
    # i가 홀수이고, value가 50 이하인 경우에만 출력
    if i % 2 == 1 and value <= 50:
      print(f"{number} X {i} = {value}")
    i = i + 1



number = int(input("몇 단?: "))
gugudan(number)
```

```python
#미션 2
import random

def rsp_advanced(games):
  my_wins, computer_wins, ties = 0, 0, 0

  for game in range(1, games + 1):
    while True:
      my_num = int(input(f"{game} 번째 판 가위(0), 바위(1), 보(2) 중에 선택하세요: "))
      if my_num in (0, 1, 2):  # 0, 1, 2 중 하나를 입력했을 때만 while 루프 탈출
        break
      else:
        print("제대로 된 값을 입력하세요")

    choices = ["가위", "바위", "보"]
    my_choice = choices[my_num]  # 사용자의 선택을 숫자에서 문자로 변환
    computer_choice = random.choice(choices)  # 컴퓨터의 선택을 랜덤으로 결정

    print("나:", my_choice)  # 사용자와 컴퓨터의 선택 출력
    print("컴퓨터:", computer_choice)

    if my_choice == computer_choice:  # 비겼을 경우
      print("비겼습니다")
      ties += 1  # 무승부 횟수 증가
    elif (my_choice == "가위" and computer_choice == "보") or \
        (my_choice == "바위" and computer_choice == "가위") or \
        (my_choice == "보" and computer_choice == "바위"):  # 이겼을 경우
      print(f"{game} 번째 판 나의 승리!")
      my_wins += 1  # 나의 승리 횟수 증가
    else:  # 졌을 경우
      print(f"{game} 번째 판 컴퓨터의 승리!")
      computer_wins += 1  # 컴퓨터의 승리 횟수 증가

  print("\n--- 최종 전적 ---")
  print(f"나의 전적: {my_wins}승 {ties}무 {computer_wins}패")
  print(f"컴퓨터의 전적: {computer_wins}승 {ties}무 {my_wins}패")

# 몇 번 게임을 할 것인지 입력
games = int(input("몇 판을 진행하시겠습니까? : "))
rsp_advanced(games)

```

```python
#미션 3

def find_even_number(num1, num2):
  numbers = [i for i in range(num1, num2 + 1)]  # 입력된 두 수 사이의 모든 숫자를 리스트로 생성
  middle_index = 0

  for i in numbers:
    middle_index += 1  # 숫자가 진행될 때마다 중앙값 인덱스 증가
    if i % 2 == 0:  # 현재 숫자가 짝수인 경우
      print(f"{i} 짝수")

    if middle_index == (num2 / 2) + 0.5:  # 중앙값 인덱스와 num2 / 2가 일치하는 경우 (중앙에 해당하는 숫자인 경우)
      print(f"{i} 중앙값")


n = int(input("첫 번째 수 입력:"))
m = int(input("두 번째 수 입력:"))
find_even_number(n, m)


```

```python
#미션 4
def is_prime(num):
  if num < 2:  # 0과 1은 소수가 아님
    return False
  for i in range(2, int(num**0.5) + 1):  # 소수임을 확인하기 위해서는 2~sqrt(num)까지 나눠보면됨
    if num % i == 0:  # 나누어 떨어지면 소수가 아님
      return False
  return True  # 나누어 떨어지지 않으면 소수임

def count_prime_number(num1, num2):
  prime_num_count = 0
  for num in range(num1, num2 + 1):
    if is_prime(num):  # 소수인 경우 prime_num_count 증가
      prime_num_count += 1
  print(f"소수개수 {prime_num_count}")

n = int(input("첫 번째 수 입력:"))
m = int(input("두 번째 수 입력:"))
count_prime_number(n, m)

```
