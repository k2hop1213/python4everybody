```python
#미션 1
import random
def bs31():
  print("베스킨라빈스 써리원 게임")
  print("------------------------")
  current_num = 0
  while True:
    my_num_list = input("My turn - 숫자를 입력하세요:").split() #입력받은 숫자들은 공백을 기준으로 분리하기
    try:
      my_num_list = [int(num) for num in my_num_list]
    except :
      print("올바른 숫자를 입력해주세요.")
      continue

    if len(my_num_list) > 3 :
      print("올바른 숫자를 입력해주세요.") #숫자는 3개 이상 외칠 수 없음
      continue
    if not all(num == current_num + i for i, num in enumerate(my_num_list, start=1)):
      print("올바른 숫자를 입력해주세요.") #숫자가 연속된 숫자인지 확인, ex) 1 3 4 는 불가능함
      continue


    current_num = my_num_list[-1] #현재 숫자를 내가 외친 마지막 숫자로 저장

    if current_num >= 31 : #내가 31을 외쳤으면 컴퓨터의 승리
      print("컴퓨터 승!")
      break

    print(f"현재 숫자: {current_num}")

    computer_turn_num = random.randint(1, 3)
    computer_num_list = [i for i in range(current_num + 1, current_num + computer_turn_num + 1)] #컴퓨터는 현재 숫자의 그 다음 숫자부터 외칠 수 있음

    if computer_num_list[-1] > 31: #31이상을 외칠 수 없으므로 31까지로 스케일을 조정해야함
      computer_num_list = computer_num_list[:-1]
    try:
      current_num = computer_num_list[-1]
    except:
      computer_num_list = 31
    computer_output = ', '.join(map(str, computer_num_list))
    print(f"컴퓨터: {computer_output}")
    print(f"현재 숫자: {current_num}")

    if current_num >= 31 :
      print("나의 승!")
      break

bs31()
```

```python
#미션 2
def grader(s, a):
  correct_answers = a[:]
  scores = []

  for student_answer in s:
    name, answer = student_answer.split(",")
    answer_list = [int(i) for i in answer]

    score = sum([1 for student_digit, correct_digit in zip(answer_list, correct_answers) if student_digit == correct_digit])*10
    scores.append((name, score))

  scores.sort(key=lambda x: x[1], reverse=True)  # 점수를 기준으로 내림차순 정렬

  grade = 0
  for name, score in scores:
    grade += 1
    print(f"학생:{name} 점수:{score}점 {grade}등")

# 학생 답
s = ["김갑,3242524215",
"이을,3242524223",
"박병,2242554131",
"최정,4245242315",
"정무,3242524315"]

# 정답지
a = [3, 2, 4, 2, 5, 2, 4, 3, 1, 2]

grader(s, a)




```

```python
import random

def guess_numbers():
  number = random.randint(0, 100)
  n_try = 1 #시도 횟수
  guessed_numbers = [] #이전 숫자들을 저장할 리스트
  cmp_pre_result = 1
  while True:
    print(f"{n_try}차 시도")
    try:
      my_num = int(input("숫자를 예측해보세요: ")) #숫자 입력받기
    except ValueError:
      print("숫자를 다시 입력해주세요")
      continue

    if my_num < 0 or my_num > 100:
      print("0부터 100 사이의 숫자를 입력해주세요.")
      continue

    if n_try > 1: #시도 횟수가 1회 이상이여야 그전 숫자와 비교 가능
      cmp_pre_result = cmp_pre_num(cmp_result, my_num, guessed_numbers)

    if cmp_pre_result == -2:
      print(f"지금까지의 시도 중 정답에 가까운 최대값은 {bigger(number,guessed_numbers)}입니다.")
    if cmp_pre_result == -3:
      print(f"지금까지의 시도 중 정답에 가까운 최소값은 {smaller(number,guessed_numbers)}입니다.")

    if cmp_pre_result == 1:
      cmp_result = cmp_num(my_num, number)
      guessed_numbers.append(my_num)

    if cmp_result == 0:
      print(f"정답입니다. {n_try}차 시도만에 예측에 성공했네요!! 게임을 종료합니다.")
      break
    else:
      n_try += 1

def cmp_pre_num(cmp_result, my_num, guessed_numbers): #이전숫자와 비교하는 함수

  if my_num in guessed_numbers:
    print("이미 예측에 사용한 숫자입니다. 다시 한번 고민해보세요.")
    return -1
  if my_num > max(guessed_numbers) and cmp_result == 1:
    print(f"앞선 예측보다 큰 숫자입니다. 더 작은 수를 고민해보세요.")
    return -2
  if my_num < min(guessed_numbers) and cmp_result == -1:
    print(f"앞선 예측보다 작은 숫자입니다. 더 큰 수를 고민해보세요.")
    return -3
  return 1

def bigger(number,guessed_numbers):
  check_num_list1=[abs(number-i) for i in guessed_numbers if number - i < 0]
  max_num = min(check_num_list1) + number
  return max_num

def smaller(number,guessed_numbers):
  check_num_list1=[number-i for i in guessed_numbers if number - i > 0]
  min_num = number - min(check_num_list1)
  return min_num

def cmp_num(my_num, number): #정답 숫자와 비교하는 함수
    if my_num == number:
        return 0
    elif my_num < number:
        print("Up! 다시 한번 고민해보세요.")
        return -1
    else:
        print("Down! 다시 한번 고민해보세요.")
        return 1

guess_numbers()
```

```python

def day_function1(week):
  days = ["월", "화", "수", "목", "금", "토", "일"]
  index = days.index(week)
  for i in range(100):
    days[index % 7]
    index+=1
  return days[(index % 7) - 1]

def day_function2(month,day):

  for _ in range(100):
    day += 1
    if month == 2 and day > 28:
      month +=1
      day = 1
    if (month == 1 or month == 3 or month == 5 or month == 7 or month == 8 or month == 10 or month == 12) and day > 31:
      month +=1
      day = 1
    elif (month == 4 or month == 6 or month == 9 or month == 11) and day > 30 :
      month +=1
      day = 1

  return month,(day-1)

def after_100(month, day, week):
  new_week = day_function1(week)  # 반환값을 변수에 저장
  new_month, new_day = day_function2(month, day)  # 반환값을 변수에 저장
  print(f"이로부터 100일 뒤의 날짜는 {new_month}월 {new_day}일 {new_week}요일입니다.")


month = int(input("월을 입력하세요:"))
day = int(input("일을 입력하세요:"))
week = input("요일을 입력하세요:")
print(f"우리의 기념일은 {month}월 {day}일 {week}요일입니다")
after_100(month, day, week)

```
